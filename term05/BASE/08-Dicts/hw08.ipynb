#### Задача 1. 

Напишите функцию, которая будет принимать числа от 1 до 10 и заменять их на слова: 1 = один, 2 = два и т.п. 

```python
def num_to_words():
    n = int(input('Введите число: '))
    while 0 < n < 11:
        print(f'{n}: {nums[n]}')
        n = int(input('Введите число: '))
    print('Число вне диапазона')

nums = {1: 'один', 2: 'два', 3: 'три', 4: 'четыре', 5: 'пять', 6: 'шесть', 7: 'семь', 
        8: 'восемь', 9: 'девять', 10: 'десять'}

num_to_words()
```

#### Задача 2. 

Пользователь вводит две строки (в каждой строке какое-то слово). Напишите функцию, которая будет проверять, что из букв строки 1 можно составить слово в строке 2, причем количество букв тоже нужно учитывать: нельзя составить слово "мама" из слова "марокко". Функция должна возвращать True или False. Мы делали такую задачу с помощью методов строк, а теперь давайте попробуем с помощью словарей.

```python
def words_construction(s1, s2):
    d1 = {sym: s1.count(sym) for sym in set(s1)}
    d2 = {sym: s2.count(sym) for sym in set(s2)}
    # print(d1, '\n', d2)
    for letter, amount in d2.items():
        if letter not in d1 or amount > d1[letter]:
            return False
    return True


user_text_1, user_text_2 = input(), input()
if words_construction(user_text_1, user_text_2):
    print(f'Из слова "{user_text_1}" можно составить слово "{user_text_2}"')
else:
    print(f'Из слова "{user_text_1}" нельзя составить слово "{user_text_2}"')
```

#### Задача 3.

В одном необычном ресторане все блюда добавляются в меню только после того, как их закажут клиенты. Но ресторан не успевает записывать все заказы и просит вас помочь.

1. Напишите программу, которая будет:
   - Собирать заказы клиентов в виде словаря, где ключ — это название блюда, а значение — количество заказов этого блюда.
   - Позволять клиентам добавлять новые заказы.
   - Удалять блюдо, если его заказали по ошибке.
   - Показывать текущее меню и количество заказов каждого блюда.

2. Реализуйте следующую функциональность:
   - Если клиент заказывает уже существующее блюдо, программа должна увеличивать количество заказов.
   - Если клиент хочет отменить заказ блюда, программа должна уменьшать количество заказов. Если количество заказов становится равным нулю, блюдо должно удаляться из меню.
   - Программа завершает работу, если клиент вводит команду выход.
   - Добавьте проверку на ввод: нельзя вводить отрицательные количества.
   - Добавьте возможность изменить количество заказов, указав новое число, вместо добавления или удаления.

3. Пример работы:
  
       Добро пожаловать в "Ресторан приключений"!
       Доступные команды: добавить, удалить, меню, выход

       Введите команду: добавить
       Введите название блюда: Пицца
       Введите количество: 2
       Блюдо "Пицца" добавлено в меню с количеством 2.

       Введите команду: добавить
       Введите название блюда: Пицца
       Введите количество: 1
       Блюдо "Пицца" обновлено: теперь количество 3.

       Введите команду: меню
       Текущее меню:
       Пицца: 3

       Введите команду: удалить
       Введите название блюда: Пицца
       Введите количество: 3
       Блюдо "Пицца" удалено из меню.

       Введите команду: выход
       До свидания!
   

```python
def adding_order():
    global orders
    name = input('Введите название блюда: ')
    am = int(input('Введите количество: '))
    while am < 1:
        print('Число должно быть положительным')
        am = int(input('Введите количество: '))
    if name not in orders:
        orders[name] =  am
        print(f'Блюдо "{name}" добавлено в меню с количеством {orders[name]}.')
    else:
        orders[name] += am
        print(f'Блюдо "{name}" обновлено: теперь количество {orders[name]}.')


def deleting_order():
    global orders
    name = input('Введите название блюда: ')
    amount_to_delete = int(input('Введите количество: '))
    while amount_to_delete < 1:
        print('Число должно быть положительным')
        amount_to_delete = int(input('Введите количество: '))
    orders[name] -= amount_to_delete
    if orders[name] < 1:
        del orders[name]
        print(f'Блюдо "{name}" удалено из меню.')
    else:
        print(f'Блюдо "{name}" обновлено: теперь количество {orders[name]}.')


def changing_order():
    global orders
    name = input('Введите название блюда: ')
    am = int(input('Введите новое количество: '))
    orders[name] = am
    print(f'Блюдо "{name}" изменено: теперь количество {orders[name]}.')

print('Добро пожаловать в "Ресторан приключений"!')
print('Доступные команды: добавить, удалить, изменить, меню, выход')
orders = dict()
command = input('\nВведите команду: ').lower()
while command != 'выход':
    if command == 'добавить':
        adding_order()
    elif command == 'меню':
        print('Текущее меню: ')
        for dish, amount in orders.items():
            print(f'{dish}: {amount}')
    elif command == 'удалить':
        deleting_order()
    elif command == 'изменить':
        changing_order()
    else:
        print('Такой команды нет')
    print()
    command = input('Введите команду: ').lower()
print('До свидания!')
```

#### Задача 4.

Создадим базу данных о продажах некоторого интернет-магазина. Будем вводить данные по строкам, где каждая строка представляет собой запись вида Покупатель товар количество, где Покупатель — имя покупателя (строка без пробелов), товар — название товара (строка без пробелов), количество — количество приобретенных единиц товара, например:

    Иванов футболка 2
    Иванов тарелка 10
    Петров ручка 5
    Сидоров кружка 1
    Петров тетрадка 2

Создайте список всех покупателей, а для каждого покупателя подсчитайте количество приобретенных им единиц каждого вида товаров. Список покупателей, а также список товаров для каждого покупателя нужно выводить в лексикографическом порядке.

```python
data = dict()
people_amt = int(input())
for _ in range(people_amt):
    person = input().split()
    if person[0] not in data:
        data[person[0]] = {person[1]: int(person[2])}
    else:
        if person[1] in data[person[0]]:
            data[person[0]][person[1]] += int(person[2])
        else:
            data[person[0]][person[1]] = int(person[2])
print(data)
for pers, thing in sorted(data.items(), key=lambda x:x[0]):
    print(pers)
    for th in sorted(thing.keys()):
        print(f'{th:>15}: {data[pers][th]}')
```
