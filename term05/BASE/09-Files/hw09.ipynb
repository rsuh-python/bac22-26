#### Задача 1.

В файле (можете взять любой файл с художественным произведением) содержится текст. Нужно сделать следующее: 

- считать текст из файла, добыть из него слова без знаков пунктуации (кто умеет регуляркой - хорошо, но в целом достаточно стандартных split и strip), собрать их во множество, чтобы не было повторяющихся слов. 
- отсортировать слова с конца, как в словаре Зализняка. 
- создать папку 'zaliznyak'
- в эту папку записать несколько файлов: каждый файл называется по букве, на которую заканчиваются слова, которые в нем содержатся. То есть, файл с названием А.txt должен содержать мама, на, окна, пара и так далее. 
- (Необязательное) Все слова должны быть выровнены по правому краю, если догадаетесь, как сделать так, чтобы выравнивание происходило по самому длинному слову - отлично. (Нужно использовать f-строки и их параметры). 

```python
# опен файл - откинуть знаки препинания сплитом и стрипом - собрать во множество
# отсортировать по последней букве как у зализняка
# создать папку зализняк
# в папку записать файлы на букву - ну словарь короче
# выровнивать по правому краю
import os 

def sorting_rule(s):
    return [ord(x.lower()) if x.lower() != 'ё' else ord('е') + 0.5 for x in s][::-1]
punct = ''',.!@?";»:()1234567890—'''
all_words = set()
with open('myfolder/the_black_cat_updated.txt', 'r', encoding='utf8') as file:
    for line in file:
        splited = [sym.strip(punct).lower() for sym in line.split() if sym.strip(punct)]
        all_words = all_words.union(splited)
# for x in all_words:
#     if len(x) == 1:
#         print(x)
# print(all_words)
all_words = sorted(all_words, key=sorting_rule)
print(all_words)
os.makedirs('homework/01dictionary/zaliznyak',exist_ok=True)
# теперь разбиваем слова по файлам
data = []
for ind, word in enumerate(all_words):
    curr_letter = word[-1]
    if ind == 0:
        ind += 1
    ind -= 1
    file_letter = all_words[ind][-1]
    if file_letter == curr_letter:
        data.append(word)
    else:
        # path = 
        with open('homework/01dictionary/zaliznyak/' + file_letter.upper() + '.txt', 'w', encoding='utf8') as dictionary:
            text = ''
            for sym in data:
                text += f'{sym:>{max(len(x) for x in data)}}'
                text += '\n'
            dictionary.write(text)
        data = []
        data.append(word)
```

Напишем программу - телефонный справочник. 

В программе должно быть:

```python
import pickle
import os.path

def download():
    with open('homework/03telephone_book.pkl', 'rb') as tel_book:
        data = pickle.load(tel_book)
        print(data) #для проверки
    return data

def search_per():
    # скачиваем словарь, инпут, через гет! чтобы не вывалилась ошибка
    data = download()
    person = input('Введите имя контакта')
    res = []
    for per in data.keys():
        if person in per:
            res.append(per)
    if not res:
        print('Такого контакта нет в вашем справочнике')
    else:
        print('Результаты поиска:')
        for per in res:
            print(f'\t{per:<{max(len(x) for x in res)}} : {data[per]}')

    
def search_num():
    data = download()
    num_search = int(input('Введите номер телефона'))
    res = []
    for per, num in data.items():
        if num_search == num:
            res.append(per)
    if not res:
        print('Такого номера нет в вашем справочнике')
    else:
        print('Результаты поиска:')
        for per in res:
            print(f'\t{per:<{max(len(x) for x in res)}} : {data[per]}')


def create(): # записываем контакт + проверка на наличие ключа
    name = input('Введите имя контакта')
    data = download()
    if name in data.keys():
        print('Такой контакт уже есть. Вы хотите его перезаписать? да/нет')
        if input() == 'да':
            print('Контакт перезаписывается')
            num = int(input('Введите новый номер телефона'))
        else:
            return
    else:
        num = int(input('Введите номер телефона'))
    data[name] = num
    save(data)
    print('Контакт записан!')
    
def save(data):
    with open('homework/03telephone_book.pkl', 'wb') as tel_book:
        pickle.dump(data, tel_book)

#проверка на наличие файла + creating
check = os.path.exists('homework/03telephone_book.pkl')
if not check:
    with open('homework/03telephone_book.pkl', 'wb') as tel_book:
        data = {}
        pickle.dump(data, tel_book)

while True:
    print('Выберите, что выхотите\n\t0 - Выйти из справочника\n\t1 - Поиск по контакту\n\t2 - Поиск по номеру телефона\n\t3 - Записать новый контакт')
    user = int(input())
    if user == 0:
        break
    elif user == 1:
        search_per()
    elif user == 2:
        search_num()
    elif user == 3:
        create()
    else:
        print('Такой команды нет')
```

#### Задача 4. 

Пользователь вводит путь к файлу, а ваша программа должна 1) проверить, что путь введен правильно и такой файл существует 2) попытаться открыть этот файл 3) считать из него текст и вывести на экран первые 200 символов, если текст длиннее 200 символов, и весь текст, если не длиннее. 

```python
import os
user_path = input('Введите путь к файлу')
existence = os.path.exists(user_path)
while not existence:
    print('Ошибка. Введите путь заново')
    user_path = input('Введите путь к файлу')
    existence = os.path.exists(user_path)
with open(user_path, 'r', encoding='utf8') as file:
    sum = 0
    text = ''
    for line in file:
        sum += len(line)
        if sum < 200:
            text += line
        elif sum - len(line) < 200:
            text += line[:200 - sum]
        else:
            break
print(text)
# print(len(text))
```
