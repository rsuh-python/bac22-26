1. Нужно открыть текстовый файл и считать из него текст в виде списка слов без пунктуации.

```python
punct = ''',.!@?";:()'''
wordlist = []
with open('files/01-1.txt', 'r', encoding='utf8') as f:
    for line in f:
        words = [w.strip(punct).replace('@', '') for w in line.strip().split()]
        wordlist.extend(words)
wordlist
```

2. Нужно записать получившийся список слов по слову на строчке в новый файл. 

```python
with open('files/words.txt', 'w', encoding='utf') as file:
        for word in wordlist:
                print(word, file=file)
```

```python
import json

all_texts = []
with open('files/13.json', 'r', encoding='utf8') as data:
    base = json.load(data)
    #в словаре ключ completions, 0 элемент списка, словарь ключ result, 0 element of list, dict key "value", dict key "choices", list  - "author"
    #извлекаем текст:
    #dict - key "data", dict - key "text"
    for ind in range(len(base)):
        if base[ind]['completions'][0]['result'][0]['value']['choices'][0] == 'author':
            all_texts.append(base[ind]['data']['text'])
for text in all_texts:
    print(text)
```
